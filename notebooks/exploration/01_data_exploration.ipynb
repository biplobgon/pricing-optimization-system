# Retail Price Optimization — Data Exploration
This notebook explores the raw dataset used by the pricing optimization system. It covers data loading, structure inspection, basic exploratory data analysis, cleaning checks, and feature engineering previews.


```python
from pathlib import Path
import sys

repo_root = Path.cwd()
for _ in range(6):
    if (repo_root / 'pyproject.toml').exists() or (repo_root / 'README.md').exists():
        break
    repo_root = repo_root.parent

sys.path.insert(0, str(repo_root / 'src'))
print(f'Repository root: {repo_root}')
print(f"Source path added: {repo_root / 'src'}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

## 1. Load the raw dataset
We use the project `DataLoader` utility so the notebook follows the same ingestion patterns used by the rest of the system.


```python
from data_ingestion.data_loader import DataLoader

loader = DataLoader(repo_root / 'data' / 'raw')
df = loader.load_raw_data('retail_price.csv')

df.head()
```

```python
print('Dataset shape:', df.shape)
print('')
print('Columns:')
print(df.columns.tolist())
print('')
print('Data types:')
print(df.dtypes)
```

## 2. Dataset overview
This section summarizes the dataset structure, missing values, and categorical distributions.


```python
print('Missing values per column:')
print(df.isnull().sum())
print('')
print('Duplicate rows:', df.duplicated().sum())

numeric_summary = df.describe(include='number').T
numeric_summary['missing'] = df.isnull().sum()
numeric_summary
```

```python
for col in ['category', 'seasonality']:
    if col in df.columns:
        print(f'\nValue counts for {col}:')
        print(df[col].value_counts(dropna=False))
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(14, 5))
sns.histplot(df['avg_price'], ax=axes[0], kde=True, color='tab:blue')
axes[0].set_title('Average Price Distribution')

sns.histplot(df['demand'], ax=axes[1], kde=True, color='tab:green')
axes[1].set_title('Demand Distribution')
plt.tight_layout()
```

## 3. Data quality checks
We inspect missing data, duplicates, and any immediate issues before feature engineering.


```python
missing_report = df.isnull().sum().rename('missing_values').to_frame()
missing_report['percent_missing'] = missing_report['missing_values'] / len(df) * 100
missing_report
```

```python
if 'category' in df.columns:
    df['category_clean'] = df['category'].astype(str).str.strip().str.title()

if 'seasonality' in df.columns:
    df['seasonality_clean'] = df['seasonality'].astype(str).str.strip().str.title()

print(df[['category', 'category_clean', 'seasonality', 'seasonality_clean']].head())
```

## 4. Cleaning preview
The `DataCleaner` utility will handle missing values and can be extended to remove outliers or normalize key fields.


```python
from data_ingestion.data_cleaner import DataCleaner

cleaner = DataCleaner(missing_strategy='mean')
df_clean = cleaner.handle_missing_values(df)

print('Missing values after cleaning:')
print(df_clean.isnull().sum())
print('')
print('Cleaning report:')
print(cleaner.get_cleaning_report(df_clean))
```

## 5. Feature engineering preview
We create time-based and price-related features using the project `FeatureEngineer` utility.


```python
from data_ingestion.feature_engineer import FeatureEngineer

engineer = FeatureEngineer(time_column='month_year', target_column='demand')
df_features = engineer.create_all_features(df_clean)

df_features.head()
```

```python
print('New features added:')
new_cols = [col for col in df_features.columns if col not in df.columns]
print(new_cols)

df_features.loc[:, ['month_year', 'year', 'month', 'quarter', 'margin', 'margin_percent', 'price_diff', 'price_ratio', 'revenue']].head()
```
